## 第1题 绘制神经元网络中常见激活函数的图形。

1. sigmod函数  
 $\sigma (x) = \frac{1}{1+e^{-x}}$
2. tanh函数  
 $\tanh (x) = \frac{e^{x}- e^{-x}}{e^{x}+ e^{-x}} $
3. ReLU函数  
 $ReLU(x) = max(0,x)$
4. softplus函数  
 $softplus(x) = log(1+e^x)$

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

```

```python
def sigmod(x):
    return 1 / (1 + np.exp(-np.array(x)))

def tanh(x):
    return (np.exp(np.array(x)) - np.exp(-np.array(x)))/(np.exp(np.array(x)) + np.exp(-np.array(x)))

def relu(x):
    return np.maximum(0, np.array(x))

def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
x = np.linspace(-6, 6, 100)
y1 = sigmod(x)
y2 = tanh(x)
y3 = relu(x)
y4 = softplus(x)
plt.plot(x, y1, '-', label='sigmod')
plt.plot(x, y2, '--', label='tanh')
plt.plot(x, y3, '-.', label='ReLU')
plt.plot(x, y4, ':', label='softplus')

plt.xlim([-6.5, 6.5])
plt.ylim([-1.5, 1.5])
plt.legend()
a=plt.gca()
a.spines['right'].set_visible(False)
a.spines['top'].set_visible(False)
a.spines['left'].set_position(('data', 0))
a.spines['bottom'].set_position(('data', 0))
```

## 第2题 绘制swish函数图形
$swish(x) = x\sigma(\beta x)$
绘制当$\beta = 0,0.5,1,100$时的图形，其中$\sigma (x) = \frac{1}{1+e^{-x}}$.

```python
betas = [0,0.5, 1, 100]
x = np.linspace(-4, 4, 100)
linestyles = ['-g', '--c', '-.m', '-r']
labels = [r'$\beta = 0$', r'$\beta = 0.5$', r'$\beta = 1$', r'$\beta = 100$']
for beta, linestyle, label in zip(betas, linestyles, labels):
    y = x * sigmod(x*beta)
    plt.plot(x, y, linestyle, label=label)

plt.xlim([-4.5, 4.5])
plt.ylim([-2.5, 4.5])
plt.legend()
a=plt.gca()
a.spines['right'].set_visible(False)
a.spines['top'].set_visible(False)
a.spines['left'].set_position(('data', 0))
a.spines['bottom'].set_position(('data', 0))
```
